```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Define the parameters
zeta5 = 0.59
zeta6 = 8
zeta7 = 0.41
zeta8 = 5

# Define the PDF function
def pdf(y):
    return zeta5 * np.exp(-zeta6 * y) + zeta7 * np.exp(-zeta8 * y)

# Compute the probability of waiting between 2 and 4 hours
probability, _ = quad(pdf, 2, 4)

print("Probability of waiting between 2 and 4 hours:", probability)

# Define the range of waiting times (in minutes)
waiting_times = np.linspace(0, 240, 1000)

# Calculate the PDF values
pdf_values = pdf(waiting_times / 60)  # Convert hours to minutes

# Plot the probability density function graph
plt.plot(waiting_times, pdf_values)
plt.xlabel('Waiting Time (minutes)')
plt.ylabel('Probability Density')
plt.title('Probability Density Function of Waiting Time to Hear Owl')
plt.grid(True)
plt.show()

# Create a histogram by the minute
plt.hist(np.random.exponential(scale=1/zeta6, size=10000), bins=100, density=True, alpha=0.5, label='Simulated data')
plt.xlabel('Waiting Time (minutes)')
plt.ylabel('Frequency')
plt.title('Histogram of Waiting Time to Hear Owl')
plt.grid(True)
plt.show()

# Compute mean, variance, and quartiles
mean = 1 / (zeta6 * 60)  # Convert mean from hours to minutes
variance = 1 / (zeta6 ** 2 * 60)  # Convert variance from hours to minutes
quartiles = np.percentile(np.random.exponential(scale=1/zeta6, size=10000), [25, 50, 75])

print("Mean waiting time:", mean, "minutes")
print("Variance of waiting time:", variance, "minutes^2")
print("Quartiles of waiting time:", quartiles, "minutes")
```
