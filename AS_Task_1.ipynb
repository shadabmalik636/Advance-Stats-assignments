```python
import matplotlib.pyplot as plt

# Proportions
proportions = [0.59, 0.41]
labels = ['For', 'Against']

# Plot
plt.bar(labels, proportions, color=['blue', 'red'])
plt.xlabel('Vote Outcome')
plt.ylabel('Proportion')
plt.title('Proportion of "For" and "Against" Votes')
plt.ylim(0, 1)  # Set y-axis limits
plt.show()
```
